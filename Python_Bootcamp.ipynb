```python
a=int(input("Enter the value of a:"))
b=int(input("Enter the value of b:"))

sum=a+b
sub=a-b
mul=a*b
div=a/b
intdiv=a//b
mod=a%b

print(f"Sum: {sum}, Subtraction: {sub}, Multiplication: {mul}, Division: {div}, Integer Division: {intdiv}, Modulus:{mod}")
print("Sum: {0}, Subtraction: {1}, Multiplication: {2}, Division: {3}, Integer Division: {4}, Modulus:{5}".format(sum,sub,mul,div,intdiv,mod))
print("Sum: %d, Subtraction: %d, Multiplication: %d, Division: %5.3f, Integer Division: %d, Modulus:%d"%(sum,sub,mul,div,intdiv,mod))
```

%d--INTEGER

%f--FLOAT   
(%P.pf --> P=padding (minimum number of blocks to be assigned for printing value) p=precision of decimal positions)

%s--STRING

```python
print("{0:^5}|{1:>4}|{2:>4}|{3:>4}|{4:>4}|{5:>4}".format("Sum","Sub","Mul","Div","Int-Div","Mod"))
print("{0:>5}|{1:>4}|{2:>4}|{3:>4}|{4:^4}|{5:>4}".format(sum,sub,mul,div,intdiv,mod))
```

{n:sp}

n=Number in list of variables which are to be written in format .
(in above example consider [sum,sub,mul,div,intdiv,mod] as a list.)

s=Symbol .
(>RIGHT ALLIGNMENT) (<LEFT ALIGNMENT) (^CENTER ALIGNMENT)

p=Padding quantity .
(minimum number of blocks to be occupied and can extent beyond that depending on string and number size.)


```python
a="champ"
print(a[3]) #identifing the element based on index
print(a[2:]) #Grab the remaing elements except upto the Index
print(a[:2]) #Grab the elements upto the Index
print(a[-1]) #Grab the Last element
print(a[:-1]) #Grab the elements except last element
print(a[::2]) #Grab everything with2steps
print(a[::-1]) #Print string backwards
```
